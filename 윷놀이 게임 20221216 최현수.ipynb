## 과제3: 단순한 윷놀이 게임 이야기
### 인공지능공학과 1학년
#### 20221216 최현수

```python
import random

com = 0
me = 0

while com<30 and me<30:  #둘중에 하나라도 30점이 넘으면 반복문이 종료 

    while com<30 : 
        compoint = random.randint(1,5)
        if compoint == 1:
            print('상대는 도를 던졌습니다')
        elif compoint == 2:
            print('상대는 개를 던졌습니다')
        elif compoint == 3:
            print('상대는 걸을 던졌습니다')
        elif compoint == 4:
            print('상대는 윷을 던졌습니다')
        else: 
            print('상대는 모를 던졌습니다')
        com += compoint
        print(compoint,'점을 획득하여 상대방의 현재 점수는', com, '점입니다')
        if compoint >= 4 : 
            print("한번 더 던질 기회가 주어집니다.")
        else : 
            print('이제 나의 차례입니다')
            break
    if com>30 : break
    while me<30 : 
        mepoint = random.randint(1,5)
        if mepoint == 1:
            print('나는 도를 던졌습니다')
        elif mepoint == 2:
            print('나는 개를 던졌습니다')
        elif mepoint == 3:
            print('나는 걸을 던졌습니다')
        elif mepoint == 4:
            print('나는 윷을 던졌습니다')
        else:
            print('나는 모를 던졌습니다')
        me += mepoint
        print(mepoint,'점을 획득하여 나의 현재 점수는', me, '점입니다') 
        if mepoint >=4 :
            print("한번 더 던질 기회가 주어집니다.")
        else:
            print('이제 상대 차례입니다')
            break
    if me>30 : break
        
if me>30:
    print('내가', me,  '점으로 이겼습니다! 상대의 점수는', com, '점 입니다')
else:
    print('상대가', com, '점으로 이겼습니다! 나의 점수는', me, '점 입니다')
```

#### 과제 후기
while반복문을 1번 사용하여 나타내기에는 너무 복잡하였고 com>=30,me>=30으로 설정하고 코드를 짜니 30점이 넘었는데도 코드는 계속 실행되었습니다....
