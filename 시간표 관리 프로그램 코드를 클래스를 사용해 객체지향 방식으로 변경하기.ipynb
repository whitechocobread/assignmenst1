```python
import pickle
import os

class Timetables() : 
    def __init__(self) :
        global Timetable
        Timetable={'월':[' ' for i in range(8)], '화':[' ' for i in range(8)], '수':[' ' for i in range(8)], '목':[' ' for i in range(8)], '금':[' ' for i in range(8)]}
        while True :
            Timetables.table()
            
    def save(txt):
        if txt==1:
            with open ('timetable.txt', 'w') as file:
                for i in Timetable.values():
                    file.writelines(i)
                    file.write('\n')





        if txt==2:
            with open('timetable.pkl','wb') as file:
                pickle.dump(Timetable,file)

    def load(txt):
        if txt==1:
            p = ['월','화','수','목','금']
            with open ('timetable.txt', 'r') as file:
                i = 0
                for line in file:
                    Timetable[p[i]] = list(line.strip('\n'))
                    i += 1




        if txt==2:
            with open('timetable.pkl','rb') as file:
                Timetable = pickle.load(file)


    def dumcheck(a,b,c):
        overlap = False
        for i in range(int(c)):
            if Timetable[a][int(b)+i-1]!= ' ':
                overlap = True
        return overlap

    def table() : 
        user=(input("1번은 시간표 등록\n2번은 시간표 저장\n3번은 시간표 불러오기\n"))
        if user=='1':
            overlap = False

            a,b=input(">>>등록할수 수업의 요일과 교시를 입력하세요 : ").split(' ')

            c=(input('>>> 등록할 수업의 연강 시간을 입력하세요:'))
            d=(input('>>> 등록할 수업의 알파벳을 입력하세요:'))


            if int(b) <= 0 or int(b)+int(c)-1> 8:
                print(">>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!")
                return


            if Timetables.dumcheck(a,b,c):
                print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!')
                return




            for i in range(int(c)):
                Timetable[a][int(b)+i-1] = d

            print('[TimeTable]\n-------------\n| |M|T|W|T|F|\n-------------')
            for i in range(8):
                print('l'+str(i+1), end='')
                for j in Timetable.keys():
                    if (Timetable[j][i] == ' '):
                        print('l'+str(Timetable[j][i]), end=" ")
                    else:
                        print('l'+str(Timetable[j][i]), end="")
                print()

        elif user=='2':
            user=(input('1.텍스트 저장, 2.피클로 저장, 3.모두 저장 '))
            if user=='1':
                Timetables.save(1)
            elif user=='2':
                Timetables.save(2)
            elif user=='3':
                Timetables.save(1)
                Timetables.save(2)

        elif user=='3':
            user=(input('1.텍스트 불러오기, 2.피클로 불러오기 '))
            if user=='1':
                file = 'timetable.txt'
                if not os.path.isfile(file):
                    print('불러올 시간표가 없습니다')
                    return
                Timetables.load(1)
            elif user=='2':
                file = 'timetable.pkl'
                if not os.path.isfile(file):
                    print('불러올 시간표가 없습니다')
                    return
                Timetables.load(2)


Timetables()
```
