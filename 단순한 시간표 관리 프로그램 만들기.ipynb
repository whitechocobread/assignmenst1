### 인공지능공학과
#### 20221216 최현수

```python
Timetable=[['']*5 for i in range(8)]

while True:
    overlap = False
    
    a,b=input(">>>등록할수 수업의 요일과 교시를 입력하세요 : ").split(' ')
    if a=='월':
        a=0
    elif a=='화':
        a=1
    elif a=='수':
        a=2
    elif a=='목':
        a=3
    elif a=='금':
        a=4

    c=(input('>>> 등록할 수업의 연강 시간을 입력하세요:'))
    d=(input('>>> 등록할 수업의 알파벳을 입력하세요:'))
    
    
    if int(b) <= 0 or int(b)+int(c)-1> 8:
        print(">>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!")
        continue
    
    
    for i in range(int(c)):
        if Timetable[int(b)+i-1][a] != '':
            overlap = True

    if overlap:
        print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!')
        continue
        
        
    
    
    for i in range(int(c)):
        Timetable[int(b)+i-1][a] = d

    print('[TimeTable]\n-------------\n| |M|T|W|T|F|\n-------------')
    for i in range(8):
        print('l'+str(i+1), end='')
        for j in range(5):
            if (Timetable[i][j] == ''):
                print('l'+str(Timetable[i][j]), end=" ")
            else:
                print('l'+str(Timetable[i][j]), end="")
        print()
```
